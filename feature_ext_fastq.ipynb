```python
from Bio import SeqIO

def extract_features_from_fastq(fastq_file):
    # Open the FASTQ file
    no_of_sequence=0
    with open(fastq_file, "r") as handle:
        # Parse the FASTQ file
        for record in SeqIO.parse(handle, "fastq"):
            # Extract features
            sequence_id = record.id
            sequence = str(record.seq)
            quality_scores = record.letter_annotations["phred_quality"]

            # Print the extracted features
            print(f"Sequence ID: {sequence_id}")
            print(f"Sequence: {sequence}")
            print(f"Quality Scores: {quality_scores}")
            no_of_sequence+=1
            print("\n")
        print(f"No of Sequences : {no_of_sequence}")

# Example usage
fastq_file = "/content/drive/MyDrive/Colab Notebooks/ERR101899_1.fastq"  # Replace with the path to your FASTQ file
extract_features_from_fastq(fastq_file)
```
